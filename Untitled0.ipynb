```python
marks = [45, 60, 72, 38, 90]

total = 0

for i in range(len(marks)):
    total += marks[i]

average = total / len(marks)

if average > 50:
    print("Class passed")
else:
    print("Class failed")
print("Average:",average)
```

```python
import pandas as pd
```

```python
data ={"student_id":[101,102,103,104,105,106],
       "attendence_percent":[92,67,81,45,74,88],
       "assignment_score":[18,12,15,8,14,19],
       "quiz_score":[72,48,65,30,55,80],
       "lab_completed":[True,False,True,False,True,True]
       }
```

```python
df = pd.DataFrame(data)
df
```

```python
df.head()
df.info()
```

```python
df.describe()
```

```python
df['total_score']=df['assignment_score']+df['quiz_score']
df
```

```python
df['eligible_student']=(df['total_score']>=75) & (df['attendence_percent']>=75) & df['lab_completed']
df
```

```python
eligible_student=df[df['eligible_student']==True]
eligible_student
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.bar(df['student_id'] ,df['attendence_percent'])

plt.plot(
    df['student_id'],
    df['attendence_percent'],
    marker='o'
)

plt.xlabel('student_id')
plt.ylabel('attendence_percentage')
plt.title('student_id vs attendence graph')

plt.show()
```
